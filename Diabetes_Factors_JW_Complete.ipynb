```python

# Day 3  Data Cleaning & EDA with Python (Diabetes Dataset)

## Step 1: Import Data Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Your Dataset
df = pd.read_csv("diabetes_dataset.csv")
df.head()

# Step 3: Inspect Data
df.info()
df.describe()
df.isnull().sum()

# Step 4: Data Cleaning
## Fill missing numeric values with mean
df.fillna(df.mean(), inplace=True)

## Remove duplicates
df.drop_duplicates(inplace=True)

## Rename columns for clarity
df.rename(columns={
    "BMI": "Body_Mass_Index"
}, inplace=True)

# Step 5: Exploratory Data Analysis
## Distribution of Age
sns.histplot(df["Age"], bins=15, kde=True)
plt.title("Age Distribution")
plt.show()

## Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

## Average Glucose by Outcome
sns.barplot(x="Outcome", y="Glucose", data=df, ci=None)
plt.title("Average Glucose by Diabetes Outcome")
plt.show()

# Step 6: Save Cleaned Data
df.to_csv("diabetes_dataset_cleaned.csv", index=False)

# Step 7: Document Insights
print("Dataset cleaned. Key patterns identified. Ready for portfolio.")
```
